```python
import numpy as np 
import matplotlib.pyplot as plt 
class Graph():
    def RandomArrayGraph():
        x_axis = np.random.randint(10,90,12)
        y_axis = np.random.randint(50,120,12)
        print(x_axis,type(x_axis),len(x_axis))
        print(y_axis,type(y_axis),len(y_axis))
        plt.plot(x_axis,marker="D",mfc="yellow",mec="black",color="red",linestyle="-.",label="first line")
        plt.plot(y_axis,marker="o",mfc="red",mec="black",color="green",linestyle="solid",label="second line")
        plt.grid(color="green")
        plt.xlabel("Market Cap")
        plt.ylabel("Share Prices")
        plt.title("Stock Market")
        plt.legend()
        plt.show()
Graph.RandomArrayGraph()
```
